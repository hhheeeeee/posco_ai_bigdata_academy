### 구동방식에 대한 counts, percents, cumulative counts, cumulative percents

```python
ds = pd.read_csv("mycars.csv")
car = ds[['automatic','model']]
count = car['automatic'].value_counts().sort_index()
cumcnt = np.cumsum(count)
percent = count / sum(count) *100
cumpct = np.cumsum(percent)
count_data = pd.DataFrame({'Count ': count, 'Cumcnt' : cumcnt, 'Percent':percent,'Cumpct':cumpct})
count_data.columns.name = 'Automatic'
count_data
```

### 모델에 대한 counts, percents, cumulative counts, cumulative percents

```python
count = car['model'].value_counts().sort_index()
cumcnt = np.cumsum(count)
percent = count / sum(count) *100
cumpct = np.cumsum(percent)
count_data = pd.DataFrame({'Count ': count, 'Cumcnt' : cumcnt, 'Percent':percent,'Cumpct':cumpct})
count_data.columns.name = 'Model'
count_data
```

### 확률 계산 실습 p.231

```python
mu = 175
sigma = 5
x = 180

# 180까지 누적확률 180이하 값
prob = stats.norm.cdf(x,mu,sigma)
print("키가 {0}이상일 경우 {1:.1f}%범위에 위치".format(x,(1-prob)*100))
```

### 확률분포 실습 p.233

```python
mu = 12
sigma = 3

x1 = 9
x2 = 15

# prob1 : 누적확률 9 이하 값 계산
prob1 = stats.norm.cdf(x1, mu ,sigma)


# prob2 : 누적확률 15 이하 달릴 확률
prob2 = stats.norm.cdf(x2, mu, sigma)

answer1 = prob2 - prob1
print("1리터를 가지고 9km이상 15km이하 달릴 확률 :{0}\n".format(answer1.round(3)))

answer2 = 1 - prob2
print("1리터를 가지고 15km이상 달릴 확률 :{0}".format(answer2.round(3)))

```

### p.243

### p. 247

```python
f = 1.5
dfnum = 10
dfden = 100

# 누적확률값 계산
prob = stats.f.cdf(x = f, dfn = dfnum, dfd = dfden)
print(prob.round(3))
```
