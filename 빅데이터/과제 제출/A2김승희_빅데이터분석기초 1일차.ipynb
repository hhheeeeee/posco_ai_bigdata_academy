### 데이터 선택 실습 p.36

```python
#데이터 구성하기
from numpy.random import randn
np.random.seed(1234)
df = pd.DataFrame(randn(10,4), columns = ['A','B','C','D']) #변수4개, 자료수 10개, 정규분포에서 추출
```

```python
# slice

# B열 선택
df[['B']]

# B, D열 선택
df[['B','D']]

# 4~8행 선택
df[4:9]
```

```python
#loc

# 2~8행, A,C열 선택
df.loc[2:8,['A','C']]

# A,C 열 선택
df.loc[:,['A','C']]

# 3~8행 선택
df.loc[3:8]

# 3행 선택(단일행)
df.loc[3:3]
```

```python
#iloc

# 0~6행, A열 선택
df.iloc[0:7,[0]]

# A열 선택
df.iloc[:,[0]]

# 2~6행 선택
df.iloc[2:7]

# 5행 선택(단일행)
df.iloc[5:6]
```

```python
# query

# D > 0.5 데이터 선택 - slice 표현
df.query('D > 0.5')

# B > D 데ㅣ터 선택 - slice 표현
df.query('B > D')

# A > 0.5 그리고 D < 0.0 데이터 선택
df.query('A > 0.5 & D < 0.0')

# A > 0.5 혹은 D < 0.0 데이터 선택
df.query('A > 0.5 | D < 0.0')
```

 ### 데이터 변환 실습 p.72

```python
# 데이터 구성하기
step1 = pd.merge(fitness1, fitness2)
step1_1 = step1.loc[:,['GENDER','AGEGROUP','WEIGHT']]
```

```python
#groupby, agg 를 이용한 평균값 계산
step2 = step1.groupby(['AGEGROUP','GENDER'])[['WEIGHT']].mean().reset_index()

#step2 = step1.groupby(['AGEGROUP','GENDER'])[['WEIGHT']].agg('mean').reset_index()
step2
```

```python
a = step2.pivot('GENDER','AGEGROUP','WEIGHT')
a
```

```python
b = step2.pivot('AGEGROUP','GENDER','WEIGHT')
b
```

#### 다른 방법  : as_index 사용

```python
step2_1 = step1.groupby(['GENDER','AGEGROUP'], as_index=False)[['WEIGHT']].mean()
step2_2 = step2_1.pivot('GENDER','AGEGROUP','WEIGHT')
step2_2
```
