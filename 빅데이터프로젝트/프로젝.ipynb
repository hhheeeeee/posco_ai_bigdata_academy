```python
# object형 column 제외한 수치형 column DataFrame 만들기
df_B_num = df_B.select_dtypes(exclude=['object'])

# describe 함수를 이용해 DataFrame의 각 column 통계치 출력하기
print(df_B_num.describe())

# 이상치를 구하기 위한 Q1, Q3, IQR 값 계산하기
Q1 = df_B_num.quantile(0.25)
Q3 = df_B_num.quantile(0.75)
IQR = Q3 - Q1

# 이상치 구하기
lower_bound = (Q1 - 1.5 * IQR)
upper_bound = (Q3 + 1.5 * IQR)
outliers = df_B_num[(df_B_num < lower_bound) | (df_B_num > upper_bound)].dropna(how='all', axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# boxplot으로 이상치 시각화
plt.figure(figsize=(18, 10))
sns.boxplot(data=df_B_num)

# 그래프 출력
plt.show()
```
