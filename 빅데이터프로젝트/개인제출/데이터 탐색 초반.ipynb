```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# '대분류'가 '식품'인 행과 계절 정보를 가져옵니다.
df_filtered = df[df['대분류'] == '의류'][['계절', '상품명', '매출액']]

# '계절' 정보와 가장 큰 '매출액'을 가진 상품 이름과 매출액 추출합니다.
grouped = df_filtered.groupby('계절').apply(lambda x: x.loc[x['매출액'].idxmax()][['상품명', '매출액']])

# 그래프를 그리기 위해 다음과 같은 코드를 실행합니다.
plt.figure(figsize=(10, 6))

# 각 계절별로 가장 매출액이 큰 한 상품만 그래프에 그립니다.
for i, row in grouped.iterrows():
    plt.bar(i, row['매출액'], label=row['상품명'])

# 그래프 옵션들을 설정합니다.
plt.title('의류 계절별 매출액 TOP 1')
plt.xlabel('계절')
plt.ylabel('매출액(억원)')
plt.legend()  # 범례 표시

# 그래프를 그립니다.
plt.show()


# 각 계절별로 가장 매출액이 큰 한 상품만 출력합니다.
for i, row in grouped.iterrows():
    print(i, row['상품명'])
```

```python
import pandas as pd


# 시간대 구분 함수 정의
def get_time_category(hour):
    hour = int(hour)
    if 0 <= hour < 3:
        return '심야'
    elif 3 <= hour < 6:
        return '새벽'
    elif 6 <= hour < 9:
        return '아침'
    elif 9 <= hour < 12:
        return '오전'
    elif 12 <= hour < 15:
        return '점심'
    elif 15 <= hour < 18:
        return '오후'
    elif 18 <= hour < 21:
        return '저녁'
    elif 21 <= hour < 24:
        return '야간'

# '시간대' 열 생성
df['시간대'] = df['시작시간'].apply(get_time_category)

```

```python
# 쉽표 없이 맨 앞에 있는 이름만 추출
# 쉼표(,)가 없는 행만 추출합니다.
single_name_rows = df[df['SH'].apply(lambda x: (',' not in x) if pd.notna(x) else False)]

# 추출한 행의 'SH' 열 값들을 리스트로 변환합니다.
single_name_list = single_name_rows['SH'].tolist()

# 중복값 제거하기
unique_single_name_list = list(set(single_name_list))

print(unique_single_name_list)
print(len(unique_single_name_list))
```

```python
# 'complete_prob' 열 값이 100을 초과하는 행만 추출합니다.
df_over_100 = df[df['complete_prob'] > 100]

# 대분류별로 'complete_prob' 열 값이 100을 넘는 행의 개수를 계산합니다.
result = df_over_100['대분류'].value_counts()

print(result)
```

```python
# 'complete_prob'가 100을 초과하는 행 개수 계산 함수
def calculate_ratio(group):
    count = len(group[group['complete_prob'] > 100])
    total_count = len(group)
    ratio = count / total_count
    return ratio

# '대분류'별로 'complete_prob'가 100을 초과하는 행의 비율 계산
ratio_by_category = df.groupby('대분류').apply(calculate_ratio)

# 결과 출력
print(ratio_by_category)
```

```python
# 한명의 쇼호스트에 대해서 대분류 별로 성공했는지 실패했는지 횟수를 알고 싶음


# 각 사람별로 '대분류'별 'complete_prob'가 100을 초과하는 행의 횟수와 그렇지 않은 횟수를 계산하는 데 사용되는 함수
def count_over_under_100(group):
    over_100_count = len(group[group['complete_prob'] > 100])
    under_100_count = len(group[group['complete_prob'] <= 100])
    return over_100_count, under_100_count

# 각 사람별로 '대분류'별 'complete_prob'가 100을 초과하는 행의 횟수와 그렇지 않은 횟수를 계산
result = {}
for name in unique_single_name_list:
    person_df = df[df['SH'] == name]
    result[name] = person_df.groupby('대분류').apply(count_over_under_100)

# 결과 출력
print(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font' , family = 'AppleGothic')

# 결과용 데이터프레임 생성
result_df = pd.DataFrame(columns=['Name', 'Category', 'Over_100', 'Under_100'])

# 결과 딕셔너리를 결과용 데이터프레임에 입력
for name, data in result.items():
    for category, counts in data.items():
        over_100, under_100 = counts
        result_df = result_df.append(
            {'Name': name, 'Category': category, 'Over_100': over_100, 'Under_100': under_100},
            ignore_index=True)

# 각 사람별 그래프 생성 함수
def create_individual_chart(name, data):
    sns.set(style='whitegrid')
    plt.figure(figsize=(6, 4))
    ax = sns.barplot(x='Category', y='Over_100', data=data)
    ax.set(xlabel='Category', ylabel='Count')
    ax.set_title(f'Count of Rows with Complete_Prob Over 100 for {name}')
    plt.savefig(f'result_bar_chart_{name}.png')
    plt.show()

# 각 사람별 그래프 생성
for name in unique_single_name_list:
    individual_data = result_df[result_df['Name'] == name]
    create_individual_chart(name, individual_data)
```
